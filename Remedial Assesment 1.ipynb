1. Product apa yang akan menghasilkan pendapatan paling tinggi

```python
total_penjualan_per_produk = sales_reciepts.groupby('product_id')['quantity'].sum().reset_index().astype(float)
total_penjualan_per_produk = total_penjualan_per_produk.merge(product[['product_id', 'product', 'current_retail_price_numeric', 'current_wholesale_price']], on='product_id')
total_penjualan_per_produk = total_penjualan_per_produk.rename(columns={'quantity': 'total_quantity'})
display(total_penjualan_per_produk)
```

```python
total_penjualan_per_produk['total_pendapatan_nonpromo'] = (total_penjualan_per_produk['total_quantity'] * total_penjualan_per_produk['current_retail_price_numeric']).astype(float)
total_penjualan_per_produk['total_pendapatan_promo'] = (total_penjualan_per_produk['total_quantity'] * total_penjualan_per_produk['current_wholesale_price']).astype(float)
total_penjualan_per_produk
```

```python
sorted_penjualan = total_penjualan_per_produk.sort_values('total_pendapatan_nonpromo', ascending=False).reset_index(drop=True)
display(sorted_penjualan)
```

```python
# Visualize top 10 products by total non-promo revenue
top10 = sorted_penjualan[['product', 'total_pendapatan_nonpromo']].head(10).copy()
top10 = top10.sort_values('total_pendapatan_nonpromo', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='total_pendapatan_nonpromo', y='product', data=top10, palette='viridis')
plt.xlabel('Total Pendapatan (Non-Promo)')
plt.ylabel('')
plt.title('Top 10 Produk berdasarkan Total Pendapatan')
for i, v in enumerate(top10['total_pendapatan_nonpromo'].values):
    plt.text(v + top10['total_pendapatan_nonpromo'].max()*0.01, i, f'{v:,.0f}', va='center')
plt.tight_layout()
plt.show()
```

2. Outlet di Kota mana yang paling banyak total pembelian

```python
merged = sales_reciepts.merge(sales_outlet[['sales_outlet_id', 'store_city']], on='sales_outlet_id', how='left')

city_totals = (
    merged.groupby('store_city')['quantity']
    .sum()
    .reset_index()
    .rename(columns={'quantity': 'total_pembelian'})
    .sort_values('total_pembelian', ascending=False)
    .reset_index(drop=True)
)

display(city_totals)

top = city_totals.iloc[0]
print(f"Kota dengan pembelian terbanyak: {top['store_city']}, total pembelian: {int(top['total_pembelian'])}")
```

```python
labels = city_totals['store_city']
sizes = city_totals['total_pembelian']

explode = [0.05 if v == sizes.max() else 0 for v in sizes]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode, wedgeprops=dict(width=0.5, edgecolor='w'))
plt.title('Proporsi Total Pembelian per Kota')
plt.axis('equal')
plt.show()
```

Outlet yang paling rame penjualannya

```python
# Group sales by outlet and count total transactions
outlet_sales = (
    sales_reciepts.groupby('sales_outlet_id')
    .agg({
        'transaction_id': 'nunique',  # Count unique transactions
        'quantity': 'sum',  # Total items sold
        'line_item_amount': 'sum'  # Total revenue
    })
    .reset_index()
)

# Merge with sales_outlet to get store information
outlet_performance = outlet_sales.merge(
    sales_outlet[['sales_outlet_id', 'store_city', 'Neighorhood']], 
    on='sales_outlet_id'
)

# Rename columns for clarity
outlet_performance = outlet_performance.rename(columns={
    'transaction_id': 'total_transactions',
    'quantity': 'total_items_sold',
    'line_item_amount': 'total_revenue'
})

# Sort by total transactions to see busiest outlets
outlet_performance = outlet_performance.sort_values('total_transactions', ascending=False)

# Display results
display(outlet_performance)

# Visualize the results
plt.figure(figsize=(12,6))
plt.bar(outlet_performance['Neighorhood'], outlet_performance['total_transactions'])
plt.xticks(rotation=45, ha='right')
plt.title('Number of Transactions by Store Location')
plt.ylabel('Number of Transactions')
plt.tight_layout()
plt.show()
```

Outlet yang paling sepi penjualannya

```python
# Temukan outlet yang paling sepi berdasarkan jumlah transaksi
least_busy = outlet_performance.sort_values('total_transactions', ascending=True).reset_index(drop=True)
display(least_busy)

# Ambil outlet dengan jumlah transaksi minimum (bisa lebih dari satu jika seri)
min_tx = least_busy['total_transactions'].min()
least_busy_outlets = least_busy[least_busy['total_transactions'] == min_tx]

print(f"Outlet paling sepi (jumlah transaksi = {int(min_tx)}):")
for _, r in least_busy_outlets.iterrows():
    print(f"- sales_outlet_id {int(r['sales_outlet_id'])}: {r['Neighorhood']} ({r['store_city']})")

# Visualisasi (urut ascending)
plt.figure(figsize=(8,4))
sns.barplot(x='total_transactions', y='Neighorhood', data=least_busy, palette='magma')
plt.title('Total Transaksi per Outlet (dari yang paling sepi ke paling ramai)')
plt.xlabel('Total Transaksi')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
